```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df=pd.read_csv('iris.csv')
df.drop(['Class'],inplace=True,axis=1)
```

```python
# Perform Agglomerative Nesting clustering
agnes = linkage(df, method='ward')
```

```python
# Plot the dendrogram for AGNES clustering
plt.title('Dendrogram (AGNES)')
dendrogram(agnes)
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
cl=cluster.fit_predict(df)
```

```python
X = df.values
plt.scatter(X[cl==0, 0], X[cl==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[cl==1, 0], X[cl==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[cl==2, 0], X[cl==2, 1], s=100, c='green', label ='Cluster 3')
plt.title('CLUSTERS AGNES')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Perform Divisive Analysis clustering
diana = linkage(df, method='single')
```

```python
# Plot the dendrogram for DIANA clustering
plt.title('Dendrogram (DIANA)')
dendrogram(diana)
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, linkage='complete')  
cl=cluster.fit_predict(df)
```

```python
X = df.values
plt.scatter(X[cl==0, 0], X[cl==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[cl==1, 0], X[cl==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[cl==2, 0], X[cl==2, 1], s=100, c='green', label ='Cluster 3')
plt.title('CLUSTERS DIANA')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
